```python
def graf (x, y, i):
    plt.scatter(x,y)
    plt.title('Distancia VS Minimo')
    plt.xlabel('Minimo')
    plt.ylabel('Distancia')
    
    plt.savefig('0'+i+'.png')
    
    B = mc.MinCua(x,y)
    b= B[0] 
    m=B[1] 
    x=B[2] 
    y=B[4] 
    
    mt = '{:.2e}'.format(m)
    bt = '{:.2e}'.format(b)
    dis = '{:.2e}'.format(0.0000103*3)
    
    #plt.show()
    plt.plot([min(x), max(x)], [m*min(x) + b, m*max(x) + b], label = str(dis) +'metros: '+str(mt)+'x+'+str(bt) )
    plt.legend()
```

```python

graf(I4, E11, '1')
graf(I4, E12, '2')
graf(I4, E13, '3')
graf(I4, E14, '4')
graf(I5, E15, '5')



#plt.savefig('analisis1.png')
```
