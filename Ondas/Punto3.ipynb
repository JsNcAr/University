```python
%pylab inline
from numpy import sin 
from numpy import cos 
from numpy import tan 
```

```python
def rs (thetai, thetat):
    return -sin(thetai-thetat)/sin(thetai+thetat)
def rp (thetai, thetat):
    return tan(thetai-thetat)/tan(thetai+thetat)

def ts (thetai, thetat):
    return ( 2*sin(thetat)*cos(thetai) )/sin(thetai+thetat)
def tp (thetai, thetat):
    return ( 2*sin(thetat)*cos(thetai) )/( sin(thetai+thetat)*cos(thetai-thetat) )

def thetat(thetai, ni, nt):
    return np.arcsin(ni*sin(thetai)/nt)
```
