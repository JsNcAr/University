```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sc
```

```python

theta_i= np.linspace(sc.pi/400,sc.pi/4,100)
theta_t=[]
ni=1
nt=1.45
```

```python
def snell():
    for j in range (len(theta_i)):
        a = math.asin((ni*math.sin(theta_i[j]))/nt)
        theta_t.append(a)
        
snell()
```

```python
def rPerpedicular():
    rPer=[]
    for j in range (len(theta_i)):
        rPer.append(-math.sin(theta_i[j]-theta_t[j])/math.sin(theta_i[j]+theta_t[j]))
    return rPer

def rParalelo():
    rPar=[]
    for j in range (len(theta_i)):
        rPar.append(math.tan(theta_i[j]-theta_t[j])/math.tan(theta_i[j]+theta_t[j]))
    return rPar
        
def tPerpendicular():
    tPer=[]
    for j in range (len(theta_i)):
        tPer.append(2*math.sin(theta_t[j])*math.cos(theta_i[j])/math.sin(theta_i[j]+theta_t[j]))
    return tPer
    
def tParalelo():
    tPar=[]
    for j in range (len(theta_i)):
        tPar.append(2*math.sin(theta_t[j])*math.cos(theta_i[j])/math.sin(theta_i[j]+theta_t[j])*math.cos(theta_i[j]-theta_t[j]))
    return tPar
```

```python
f=plt.figure(figsize=(14,14))
ax1 = f.add_subplot(321,title="Coeficiente de reflexion paralelo vs angulo de inscidencia", frameon=True, xlabel="angulo de incidencia (rad)", ylabel="Coeficiente de reflexion paralelo")
plt.plot(theta_i,rPerpedicular())
ax2 = f.add_subplot(322,title="Coeficiente de reflexion paralelo vs angulo de inscidencia", frameon=True, xlabel="angulo de incidencia (rad)", ylabel="Coeficiente de reflexion paralelo")
plt.plot(theta_i,rParalelo())
ax3 = f.add_subplot(323, title="Coeficiente de transmicion perpendicular vs angulo de inscidencia", frameon=True, xlabel="angulo de incidencia(rad)", ylabel="Coeficiente de transmicion perpendicular")
plt.plot(theta_i,tPerpendicular())
ax4 = f.add_subplot(324,title="Coeficiente de transmicion paralelo vs angulo de inscidencia",frameon=True,xlabel="angulo de incidencia (rad)", ylabel="Coeficiente de transmicion paralelo")
plt.plot(theta_i,tParalelo())

```
